```python
y = np.random.rand(50)
x = np.random.rand(50)
plt.scatter(x,y)
```

```python
clr = np.random.rand(50)
size = 1000*np.random.rand(50)
plt.scatter(x,y,c =clr,s = size,alpha = 0.5)
plt.xlabel("xaxis")
plt.ylabel("yaxis")

#alpha represents intensity
```
