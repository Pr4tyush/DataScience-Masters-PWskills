# imputation Technique
# 1 -mean value imutation

```python
sns.displot(df['age'])
```

```python
df["Age_mean"] = df['age'].fillna(df['age'].mean())
#this technique will fill tha nan with the mean value of that particular column
```

```python
df[["Age_mean","age"]]
#this technique will fill tha nan with the mean value of that particular column
```

# meadian Value Imputation

```python
df["Age_median"] = df['age'].fillna(df['age'].median())
```

```python
df[["Age_median","age"]]
```

```python
#for outlier always use the median
```
